# Case Studies

## 1. What are the total Sales, Quantity, and Profit?

```python
Sales = df["Sales"].sum().round(2)
Quantity = df["Quantity"].sum()
Profit = df["Profit"].sum().round()
```

```python
print("The Total Sales is: ",Sales)
print("The Total Quantity is: ",Quantity)
print("The Total Profit is: ",Profit)
```

### 2. Which are the Top 10 highest selling products?

```python
gr = df.groupby(["Product Name"]).agg(Qty=("Quantity","sum")).round(2)
gr = gr.sort_values(by=["Qty"],ascending= False).reset_index().head(10)
gr
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=90)
ax = sns.barplot(data = gr , x="Product Name", y = "Qty", palette = "viridis")
for i in ax.containers:
    ax.bar_label(i)
plt.title("Sales of top 10 Products")    
plt.show()
```

###  3. Which 5 categories generates the highest Profit?

```python
gr1 = df.groupby(["Category"]).agg(Total_Profit=("Profit","sum")).round(2)
gr1 = gr1.sort_values(by=["Total_Profit"],ascending= False).reset_index().head(5)
gr1
```

```python
plt.figure(figsize=(8,4))
plt.pie(gr1["Total_Profit"], labels= gr1["Category"], autopct="%1.2f%%", wedgeprops={"edgecolor":"black" , "linewidth":0.5})
plt.title("Top 5 Categories With Highest Profit")
plt.show()
```

### 4. State-wise Sales and Profit Analysis

```python
gr2 = df.groupby("State").agg({"Sales":"sum","Profit":"sum"})
gr2 = gr2.reset_index().head(5).round(2)
gr2
```

```python
plt.figure(figsize=(7,4))
sns.barplot(data=gr2, x ="State", y="Sales",palette="viridis")
plt.xticks(rotation=90)
plt.title("State Wise Sales Analysis")
plt.show()


plt.figure(figsize=(7,4))
sns.barplot(data=gr2, x ="State", y="Profit",palette="viridis")
plt.xticks(rotation=90)
plt.title("State Wise Profit Analysis")
plt.show()
```

 ## 5. Monthly Sales Trend

```python
df["Month_No"] = df["Order Date"].dt.month
df
```

```python
df["Month_Name"] = df["Order Date"].dt.month_name()
df
```

```python
df["Month"] = df["Month_No"].astype(str) + "-" + df["Month_Name"]
df
```

```python
gr = df.groupby("Month_Name").agg(Total_Sales=("Sales","sum")).round(2)
gr
```

```python
sns.lineplot(data= gr, x="Month_Name", y="Total_Sales")
plt.xticks(rotation=90)
```

### 6. Year-wise performance of Sales and Profit

```python
df["Year"] = df["Order Date"].dt.year
df
```

```python
gr = df.groupby("Year").agg(Total_Sales=("Sales","sum"), Total_Profit=("Profit","sum")).reset_index().round(2)
gr
```

```python
plt.figure(figsize=(15,6))
ax = gr.set_index("Year")[["Total_Sales","Total_Profit"]].plot(kind="bar",stacked= False)
for container in ax.containers:
    ax.bar_label(container)
plt.show()  
```

### 7. Which top 10 customers placed the highest number of orders?

```python
gr = df.groupby("EmailID").agg({"Order ID":"count"})
gr = gr.reset_index()
gr = gr.sort_values("Order ID", ascending=False)
gr = gr.head(10)
gr
```

```python
plt.figure(figsize=(10,6))
plt.bar(gr["EmailID"], gr["Order ID"])
plt.title("Top 10 Customers with Highest Number of Orders")
plt.xlabel("Customer Email")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.show()
```

### 8. Which products have negative profit (loss-making)?

```python
gr = df.groupby("Product Name").agg({"Profit":"sum"}).round(2)
gr = gr[gr["Profit"]<0]
gr
```

### 9. What is the Average Order Value (AOV)?

```python
gr = df.groupby("Order ID").agg(Total_Sales=("Sales","sum"))
gr
```

```python
print("Average Order Value is :" , gr["Total_Sales"].mean().round(2))
```

### 10. How many repeat customers do we have?

```python
gr = df.groupby("EmailID").agg(Customer_Count=("EmailID","count"))
gr = gr[gr["Customer_Count"]>1]
gr = gr.sort_values(by="Customer_Count", ascending=False).reset_index()
gr
```

###  11. Which top 10 category sells the most quantity?

```python
gr = df.groupby("Category").agg(Qty_Ordered=("Quantity","sum"))
gr = gr.reset_index()
gr = gr.sort_values("Qty_Ordered", ascending=False)
gr = gr.head(10)
gr
```

```python
plt.figure(figsize=(8,4))
plt.pie(gr["Qty_Ordered"],labels=gr["Category"],autopct="%1.2f%%",wedgeprops={"edgecolor":"black","linewidth":0.5})
plt.title("Top 10 Categories With Highest Quantity Sold")
plt.show()
```

###  12. Calculate top 10 Profit Margin % for each product

```python
gr = df.groupby("Product Name").agg({"Profit":"sum", "Sales":"sum"}).round(2)
gr["Profit_Margin"] = ((gr["Profit"]/gr["Sales"])*100).round(2)
gr = gr.sort_values("Profit_Margin", ascending=False).head(10)
gr = gr.reset_index()
gr
```

```python
plt.figure(figsize=(10,5))
colors = plt.cm.viridis(np.linspace(0,1,len(gr)))
plt.bar(gr["Product Name"], gr["Profit_Margin"], color=colors)
plt.title("Top 10 Products by Profit Margin %")
plt.xlabel("Product Name")
plt.ylabel("Profit Margin (%)")
plt.xticks(rotation=75)
plt.show()
```

# 📈 Prescriptive Analysis

 Based on the above analysis, several strategic actions can help improve the company's performance:

- **Focus on Top-Selling Products:** Increase marketing efforts and maintain strong inventory levels for the top-performing products to maximize revenue.
- **Improve Low or Negative Profit Products:** Review pricing strategies, supplier costs, or discounts for loss-making products to reduce or eliminate losses.
- **Strengthen High-Profit Categories:** Invest more in the most profitable categories through promotions and product expansion.
- **Target High-Value Customers:** Implement loyalty programs or personalized offers for customers who place frequent orders to increase repeat purchases.
- **Optimize Shipping Operations:** Reduce shipping time by improving logistics and warehouse processes to enhance customer satisfaction and potential profitability.
- **Leverage Seasonal Trends:** Plan marketing campaigns and inventory management around months with historically higher sales and profit.

Implementing these strategies can improve overall profitability, operational efficiency, and long-term business growth.
